# Performing Mann-Kendall Test on temperature data


```
import pymannkendall as mk
import matplotlib.pyplot as plt
import statsmodels.api as sm
%matplotlib inline
```

```
dfx = pd.DataFrame(d1["cdd"][70][50])
dfx.plot(figsize=(12,8));
```

```
df2 = pd.DataFrame(d2["WSDI"][60][65])
df2.plot(figsize=(12,8));
```

```
df2 = pd.DataFrame(d2["WSDI"][80][50])
df2.plot(figsize=(12,8));
```

```
df2 = pd.DataFrame(d2["WSDI"][40][50])
df2.plot(figsize=(12,8));
```

```
result = mk.original_test(df2, alpha=0.05)
print(result)
```

```
mk.sens_slope(df2)
```

```
print(result.intercept)
```

#Sample plot for a trend

```
data = df2
fig, ax = plt.subplots(figsize=(12, 8))
res = mk.hamed_rao_modification_test(data)
trend_line = np.arange(len(data)) * res.slope + res.intercept
ax.plot(data)
ax.plot(data.index, trend_line)
ax.legend(['data', 'trend line'])
```

#Results of Mann-Kendall Test for complete Temperature data

```
# Using intercept
temp_data = np.zeros([121,121])
for i in range(121):
  for j in range(121):
    df = pd.DataFrame(d2["WSDI"][i][j])
    # print(mk.original_test(df, alpha=0.05))
    res = mk.original_test(df, alpha=0.05)
    temp_data[i][j] = res.intercept
```

#Plot of Precipitation Data

```
prec_data = np.zeros([121,121])
for i in range(121):
  for j in range(121):
    df = pd.DataFrame(d1["cdd"][i][j])
    # print(mk.original_test(df, alpha=0.05))
    res = mk.original_test(df, alpha=0.05)
    prec_data[i][j] = res.intercept
```
